```python
# ordinal encoding happens for categorical variables

import pandas as pd
import numpy as np
```

```python
def MyOrdinalEncoder(df):
    # mapping each unique var with a unique number 
    for colname in df:
        unique_vals=df[colname].unique()
        val_vs_num={unique_vals[i]:i for i in range(len(unique_vals))}
        for val in unique_vals:
            df[colname].replace(val,val_vs_num[val],inplace=True)
```

```python
dict1={"hello":["hola","hi","hi","henlo","bonjour","hi","henlo","henlo","bonjour","hola","hi"]}
df=pd.DataFrame(dict1)
df
```

```python
MyOrdinalEncoder(df)
df
```
