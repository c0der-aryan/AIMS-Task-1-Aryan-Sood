```python
import pandas as pd
import numpy as np
```

```python
def SimpleImputer(df,type):
    for colname in df:
        if type=="mean":
            imputeval=np.nanmean(df[colname])
        if type=="median":
            imputeval=np.nanmedian(df[colname])
        if type=="mode":
            imputeval=np.nanmode(df[colname])
        df[colname].replace(np.nan,imputeval,inplace=True)
```

```python
dict1={"hello":[1,2,3,4,np.nan,2,5,np.nan,4,5,6]}
df=pd.DataFrame(dict1)
df
```

```python
type="mean"
SimpleImputer(df,type)
df
```
